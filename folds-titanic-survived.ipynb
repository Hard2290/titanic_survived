```python
import pandas as pd
import numpy as np
from sklearn import model_selection
```

```python
train = pd.read_csv("/kaggle/input/titanic/train.csv")
test = pd.read_csv("/kaggle/input/titanic/test.csv")
```

```python
train["kfold"] = -1
y = train["Survived"]
```

```python
kf = model_selection.StratifiedKFold(n_splits=5, shuffle=True, random_state=6)
for fold, (train_indices, valid_indices) in enumerate(kf.split(X=train, y=y)):
    train.loc[valid_indices, "kfold"] = fold
```

```python
train.to_csv("folds_titanic.csv", index=False)
```
